```python
import pandas as pd
import numpy as np
```

## Task 1 
-Making three dataframe in python and saving them as three .csv files. 

```python
#Creating Project Dataframe

project_data={
    "ID":["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project":["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8","Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost":[1002000,2000000,4500000,5500000,"",680000,400000,350000,"",300000,2000000,1000000,3000000,200000],
    "Status":["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]
}

project_dataframe=pd.DataFrame(project_data)
project_dataframe
```

```python
#Creating Employee Dataframe

employee_data={
    "ID":["A001","A002","A003","A004","A005"],
    "Name":["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender":["M","F","M","F","F"],
    "City":["Paris","London","Berlin","NewYork","Madrid"],
    "Age":[25,27,29,31,30]
}
employee_dataframe=pd.DataFrame(employee_data)
employee_dataframe
```

```python
#Creating  Seniority Level DataFrame
seniority_level_data={
    "ID":["A001","A002","A003","A004","A005"],
    "Designation Level":[2,2,3,2,3]
}

seniority_level_dataframe=pd.DataFrame(seniority_level_data)
seniority_level_dataframe
```

```python
# Saving the dataframes in csv format

project_dataframe.to_csv("project_dataframe.csv", index="False")
employee_dataframe.to_csv("employee_dataframe.csv", index="False")
seniority_level_dataframe.to_csv("seniority_level_dataframe.csv", index="False")
```

```python
# Loading project_dataframe CSV File

project_dataframe_loaded = pd.read_csv("project_dataframe.csv",index_col=0)
project_dataframe_loaded
```

```python
# Loading employee_dataframe CSV File
employee_dataframe_loaded = pd.read_csv("employee_dataframe.csv",index_col=0)
employee_dataframe_loaded 
```

```python
# Loading seniority_level_dataframe CSV File
seniority_level_dataframe_loaded = pd.read_csv("seniority_level_dataframe.csv",index_col=0)
seniority_level_dataframe_loaded
```

## Task 2 
Computing the missing values in the cost column in the Project dataframe. Replacing the missing values by using "for loop" and running average. 

```python
project_dataframe_loaded["Cost"]
```

```python
#Finding average cost value
avg_cost = round(np.mean(project_dataframe_loaded["Cost"]),2)
avg_cost
```

```python
#Checking if there is any blank value in the cost column of Project Dataframe. 
#If true then putting average cost in those blank spaces.
for i in range(len(project_dataframe_loaded)):
    if pd.isna(project_dataframe_loaded.loc[i, "Cost"]):  # check if value is NaN
        project_dataframe_loaded.loc[i, "Cost"] = avg_cost
         
```

```python
project_dataframe_loaded["Cost"] 
```

## Task 3 
Splitting the Name column in the Employee dataframe into two new columns “First Name”, 
and “Last Name” and removing the older “Name” column.

```python
#Splitting the name column into 2 new columns : "First Name" and "Last Name"
employee_dataframe_loaded[["First Name", "Last Name"]] = employee_dataframe_loaded["Name"].str.split(" ", n=1, expand=True)
employee_dataframe_loaded
```

```python
# Dropping the old 'Name' column
employee_dataframe_loaded.drop(columns=["Name"], inplace=True)
employee_dataframe_loaded
```

## Task 4 
Joining all the three dataframes in one single dataframe and naming it “Final”

```python
#First joining Project Dataframe and Employee Dataframe
first_join_dataframe = pd.merge(left=project_dataframe_loaded, right=employee_dataframe_loaded, on="ID",how="outer")
first_join_dataframe
```

```python
#Joining first_join_dataframe and Seniority Level dataframe
Final=pd.merge(left=first_join_dataframe, right=seniority_level_dataframe_loaded, on="ID", how="outer")
Final
```

### Task 5 
Adding a new bonus column in the Final dataframe and giving a 5% bonus concerning project 
cost only to employees who have finished the projects.

```python
for i in range(len(Final)):
    if Final.loc[i,"Status"]=="Finished":
        Final.loc[i,"Bonus"]=round(Final.loc[i,"Cost"]*(5/100),2)
    else:
        Final.loc[i,"Bonus"]=0


Final
    
```

#### Task 6 
Demoting the designation level by 1, whose projects have status “fail”. Also deleting the 
employees record whose designation level is above 4.

```python
#Finding the employees project status "Failed" and demoting their designation level by 1
for i in range(len(Final)):
    if Final.loc[i,"Status"]=="Failed":
        Final.loc[i,"Designation Level"]-=1

Final
```

```python
#Deleting the employees record whose designation level is above 4.
for i in range(len(Final)):
    if Final.loc[i,"Designation Level"]>4:
        print(Final.loc[i, "ID"])
        del  Final.loc[i]
```

### Task 7 
Adding “Mr.” and “Mrs.” to the first name column and dropping the gender column. 

```python
#Adding "Mr." and "Mrs." to the first name column
for i in range(len(Final)):
    if Final.loc[i,"Gender"]=="M":
        Final.loc[i,"First Name"]="Mr. "+Final.loc[i,"First Name"]
    else:
        Final.loc[i,"First Name"]="Mrs. "+Final.loc[i,"First Name"]

Final
```

```python
#Dropping gender column
Final.drop(columns=["Gender"], inplace=True)
Final
```

## Task 8 
Promoting designation level by 1 for the employees whose age is more than 29 years using 
IF condition.

```python
#Finding the employees whose age>29 and promoting their designation level by 1
for i in range(len(Final)):
    if Final.loc[i,"Age"]>29:
        Final.loc[i,"Designation Level"]+=1

Final
```

## Task 9 
Adding the cost of all projects for each Employee and saving it in a new dataframe 
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
#Grouping on the basis of "ID" and "First Name" and summing the cost
TotalProjCost= Final.groupby(["ID","First Name"])["Cost"].sum().reset_index()
TotalProjCost
```

```python
#Renaming "Cost" column as "Total Cost" column
TotalProjCost.rename(columns = {"Cost": "Total Cost"}, inplace=True)
```

```python
TotalProjCost
```

## Task 10 
Printing all the employee details whose city name contains the letter “o” in it.

```python
# Filtering employees whose city contains "o" and printing it
employees_with_o = Final[Final["City"].str.contains("o", case=False, na=False)]

employees_with_o
```
